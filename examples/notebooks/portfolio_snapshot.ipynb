# Portfolio Snapshot Workflow

Required environment variables:
- `ROBINHOOD_USERNAME`
- `ROBINHOOD_PASSWORD`

**WARNING:** Never commit notebooks with saved output containing your account data.

```python
import os
from dotenv import load_dotenv

load_dotenv()

username = os.environ.get("ROBINHOOD_USERNAME")
password = os.environ.get("ROBINHOOD_PASSWORD")
print("Credentials loaded." if username and password else "Credentials missing!")
```

```python
from open_stocks_mcp.server.app import setup_brokers
from open_stocks_mcp.tools.robinhood_account_tools import (
    get_account_info,
    get_portfolio,
    get_positions
)

# Authenticate with brokers
await setup_brokers(username, password)
```

```python
# MCP tool: account_info
info = await get_account_info()

# MCP tool: portfolio
port = await get_portfolio()

# MCP tool: positions
pos = await get_positions()

print("Account Info:", info)
print("Portfolio:", port)
print("Positions:", pos)
```
