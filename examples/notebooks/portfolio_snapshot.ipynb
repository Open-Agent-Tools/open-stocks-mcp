# Portfolio Snapshot Workflow

Required environment variables:
- ROBINHOOD_USERNAME
- ROBINHOOD_PASSWORD

This notebook reads account and portfolio state through the local MCP server. It does not place orders.

## Start the MCP server

Run this in a terminal before executing the notebook:
```bash
open-stocks-mcp-server --transport http --port 3001
```

````python
import os

from mcp import ClientSession
from mcp.client.streamable_http import streamablehttp_client

mcp_http_url = os.environ.get("MCP_HTTP_URL", "http://localhost:3001/mcp")
````

````python
async with streamablehttp_client(mcp_http_url) as (read, write, _):
    async with ClientSession(read, write) as session:
        await session.initialize()
        tools = await session.list_tools()

portfolio_tool_names = [
    tool.name
    for tool in tools.tools
    if tool.name in {"portfolio", "positions", "stock_price", "account_info"}
]

portfolio_tool_names
````

````python
async with streamablehttp_client(mcp_http_url) as (read, write, _):
    async with ClientSession(read, write) as session:
        await session.initialize()
        portfolio_result = await session.call_tool("portfolio", {})
        positions_result = await session.call_tool("positions", {})

portfolio_result, positions_result
````

````python
async with streamablehttp_client(mcp_http_url) as (read, write, _):
    async with ClientSession(read, write) as session:
        await session.initialize()
        aapl_quote = await session.call_tool("stock_price", {"symbol": "AAPL"})

aapl_quote
````
