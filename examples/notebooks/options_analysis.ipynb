# Options Analysis Workflow

Required environment variables:
- `ROBINHOOD_USERNAME`
- `ROBINHOOD_PASSWORD`

**WARNING:** Never commit notebooks with saved output containing your account data.

```python
import os
from dotenv import load_dotenv

load_dotenv()

username = os.environ.get("ROBINHOOD_USERNAME")
password = os.environ.get("ROBINHOOD_PASSWORD")
symbol = os.environ.get("OPTIONS_EXAMPLE_SYMBOL", "AAPL")
print("Credentials loaded." if username and password else "Credentials missing!")
```

```python
from open_stocks_mcp.server.app import setup_brokers
from open_stocks_mcp.tools.robinhood_options_tools import (
    find_tradable_options,
    get_options_chains,
    get_open_option_positions
)

# Authenticate with brokers
await setup_brokers(username, password)
```

```python
# MCP tool: options_chains
chains = await get_options_chains(symbol)

# MCP tool: find_options
options = await find_tradable_options(symbol, expiration_date=None, option_type="call")

# MCP tool: open_option_positions
positions = await get_open_option_positions()

print("Options Chains keys:", chains.keys() if isinstance(chains, dict) else type(chains))
print("Tradable Options:", len(options.get("options", [])) if isinstance(options, dict) else type(options))
print("Open Positions:", positions)
```
