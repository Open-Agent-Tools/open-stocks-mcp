Required environment variables:
- ROBINHOOD_USERNAME
- ROBINHOOD_PASSWORD

This notebook uses read-only market and portfolio tools. No order placement calls are included.

## Connect to the MCP server

Start the server first:
```bash
uv run open-stocks-mcp-server --transport http --port 3001
```

````python
import os

from mcp import ClientSession
from mcp.client.streamable_http import streamablehttp_client

mcp_http_url = os.environ.get("MCP_HTTP_URL", "http://localhost:3001/mcp")
````

````python
async with streamablehttp_client(mcp_http_url) as (read, write, _):
    async with ClientSession(read, write) as session:
        await session.initialize()
        aapl_quote = await session.call_tool("stock_price", {"symbol": "AAPL"})
        market_status = await session.call_tool("market_hours", {})
        portfolio_snapshot = await session.call_tool("portfolio", {})

aapl_quote, market_status, portfolio_snapshot
````
